```python
 
def monte(x,y,func,N):
    count=0.0
    x_=list()
    y_=list()
    
    for i in range(N):
        if x[i]>=-1 and x[i]<=1:
            if y[i]<=func(x[i]):
                count=count+1.0
                x_.append(x[i])
                y_.append(y[i])
    return  count/N,count,x_,y_
```

```python
import math 
def circle(radius):
    def get_y(x):
        return math.sqrt(radius**2-x**2)
    return get_y
r_1=circle(1)
```

```python
from matplotlib import pyplot as mp
import numpy as np
def Guassian(mu,sig):
    def gaussian(x):
        return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))+0.01
    return gaussian

x_values = np.linspace(-2, 2, 120)
x_value=np.linspace(-1, 1, 120)
mu, sig=0, 1
g1=Guassian(mu, sig)
 
mp.plot(x_values, g1(x_values))
r0=1.0
y_values=list(map(r_1,x_value))
mp.plot(x_value,y_values)
 
mp.show()
#from this graph, I decid to use the gaussian curve in range from -1 to 1, where circle is also fully covered in this range
```

```python
import random
def random_function_point(N,min_,max_,func):
    x=list(np.random.uniform(low=min_, high=max_, size=(N)))
    y_=list()
    for i in x:
        y=random.uniform(0,func(i))
        y_.append(y)

    return  x,y_
```

```python
x,y=random_function_point(1000,-1,1,g1)
for i in range(len(x)):
    mp.scatter(x[i] ,y[i])
```

```python
factor,count,x_,y_=(monte(x,y,r_1,N=1000))
print ('In 1000 points '+str(count)+' falls into circle')
area=factor*1.73125*2
#I obtain the area of the guassian curve from direct intergral of function above from -1 to 1
print ('pi value we got here is '+str(area))
```
