 Approximation the pi value using Monte Carlo Method
 
 The area of a circle with radius 1 is equal to pi, therefore, we can find the this area value to approach the value of pi

```python
import random
def Generater(N,min_=0,max_=100000):
    out_x=list()
    out_y=list()
    ### BEGIN SOLUTION
    for i in range (N):
         
        x=float(random.randint(min_,max_))/max_
        y=float(random.randint(min_,max_))/max_
        out_x.append(x)
        out_y.append(y)
    return out_x,out_y
 
```

```python
import math 
def circle(radius):
    def get_y(x):
        return math.sqrt(radius**2-x**2)
    return get_y
```

```python
#creat circle with radius equal 1
r_1=circle(1)
x,y=Generater(10000)
x_1,y_1=Generater(1000)
 
```

```python
import matplotlib.pyplot as plt
def monte(x,y,func,N):
    count=0.0
     
    for i in range(N):
        if y[i]<=func(x[i]):
            count=count+1.0
    return  count/N,count
```

```python
factor,count=(monte(x,y,r_1,N=10000))
area=factor*4
print ('10000 random points applies here')
print ('pi value='+str(area)) 
print (str(count)+' points falls into the cirble in 1000 random points')
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
def graph_(x,y,func):  
    r0=1.0
    for i in range(len(x)):
        if y[i]<=func(x[i]): 
            plt.scatter(x[i], y[i] , marker='^')
             
        else:
            plt.scatter(x[i], y[i] , marker='o')
    theta = np.arange(0, np.pi / 2, 0.01)
    plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))
 
```

```python
#We applied 1000 random point distributed in the square,where triangles are the point falls into the circle
graph_(x_1,y_1,r_1)
factor,count=monte(x_1,y_1,r_1,N=1000)
print ('1000 random points applies here')
print ('pi value='+str(factor*4)) 
print (str(count)+' points falls into the cirble in 1000 random points')
#since my computer takes to long to plot lager volume of data, so I only applied 1000 points here, however, the pi value i got from 10000 random points is more close to the thereotical value of pi 
```

```python
from matplotlib import pyplot as mp
import numpy as np
def Guassian(mu,sig):
    def gaussian(x):
        return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))*1.01
    return gaussian

x_values = np.linspace(-2, 2, 120)
x_value=np.linspace(-1, 1, 120)
mu, sig=0, 1
g1=Guassian(mu, sig)
 
mp.plot(x_values, g1(x_values))
r0=1.0
y_values=list(map(r_1,x_value))
mp.plot(x_value,y_values)
 
mp.show()
```

```python
def random_function_point(N,min_,max_,func):
    x=list(np.random.uniform(low=min_, high=max_, size=(N)))
    y_=list()
    for i in x:
        y=random.uniform(0,func(i))
        y_.append(y)

    return  x,y_
```
